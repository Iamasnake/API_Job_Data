### Step 5: Handling Missing Values


**Task 3: Identify and Handle Missing Values**
   1. Identify missing values for all columns in the dataset.
   2. Choose a column with significant missing values (e.g., EdLevel) and impute with the most frequent value.


```python
## Write your code here
sum_nul_df = df.isnull().sum().to_frame(name = 'Missing Values Count')  #saving the count of missing values and column names into a df 
sum_nul_df.sort_values(by='Missing Values Count', ascending = False, inplace = True) #sorting the values in descending order
for col, missing in sum_nul_df['Missing Values Count'].items():  #using a loop in order to print all column - count pairs.
    print(col, missing)
    
```

```python
import numpy as np   #so I can use the np.nan for replacing the missing values with the most common value. 
```

```python
most_common = df['EdLevel'].value_counts().idxmax()  #finding the most common value in the colummn, and assigning to a variable.
df['EdLevel'] = df['EdLevel'].replace(np.nan, most_common)  #replacing the missing values with the most common value in the column
print(df['EdLevel'].isnull().sum()) #checking that there are no more missing values in the column.
```

```python

most_common = df['CodingActivities'].value_counts().idxmax()  #finding the most common value in the colummn, and assigning to a variable.
df['CodingActivities'] = df['CodingActivities'].replace(np.nan, most_common)  #replacing the missing values with the most common value in the column
print(df['CodingActivities'].isnull().sum()) #checking that there are no more missing values in the column.
```

```python

most_common = df['LearnCodeOnline'].value_counts().idxmax()  #finding the most common value in the colummn, and assigning to a variable.
df['LearnCodeOnline'] = df['LearnCodeOnline'].replace(np.nan, most_common)  #replacing the missing values with the most common value in the column
print(df['LearnCodeOnline'].isnull().sum()) #checking that there are no more missing values in the column.
```

```python

most_common = df['WorkExp'].value_counts().idxmax()  #finding the most common value in the colummn, and assigning to a variable.
df['WorkExp'] = df['WorkExp'].replace(np.nan, most_common)  #replacing the missing values with the most common value in the column
print(df['WorkExp'].isnull().sum())
```

```python

most_common = df['Industry'].value_counts().idxmax()  #finding the most common value in the colummn, and assigning to a variable.
df['Industry'] = df['Industry'].replace(np.nan, most_common)  #replacing the missing values with the most common value in the column
print(df['Industry'].isnull().sum())
```
