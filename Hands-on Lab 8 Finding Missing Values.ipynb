### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
sum_null_df = df.isnull().sum().to_frame(name = 'Missing Values') #creating a df of the missing values and their count
sum_null_df.sort_values(by= 'Missing Values', ascending = False, inplace = True) #sorting the df by descending count
for col, missing in sum_null_df['Missing Values'].items(): #looping trough all values and their coount and printing them 
    print(col, missing)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
plt.figure(figsize=(12, 8))
sns.heatmap(sum_null_df, cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
print(df['TBranch'].isnull().sum())
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
import numpy as np
most_common = df['TBranch'].value_counts().idxmax()
df['TBranch'] = df['TBranch'].replace(np.nan, most_common)
print(df['TBranch'].isnull().sum())
```
